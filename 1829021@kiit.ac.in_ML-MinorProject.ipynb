# Relations between gender and math,reading,writing scores

```python
student.groupby('gender').mean().plot(kind='bar')
```

Graph shows female gender performs better than male gender

```python
student.plot(kind='scatter',x='gender',y='math score')
plt.show()
student.plot(kind='scatter',x='gender',y='reading score')
plt.show()
student.plot(kind='scatter',x='gender',y='writing score')
plt.show()
```

The above graph scatter the math,reading,writing score point on x axis having two values Male and Female.

```python
sns.set_style("whitegrid")
sns.pairplot(student,hue='gender',height=3)
plt.show()
plt.close()
```
